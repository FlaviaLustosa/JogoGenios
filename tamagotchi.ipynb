```python
import numpy as np
from IPython.display import display
from PIL import Image
import time # necessário para contar o tempo

#variaveis
animal_vivo=1
animal_dict={}
alimentacao=['racao', 'fruta', 'legume','carne']
lista_desejo=['Comer', 'Dormir', 'Brincar','Banho','Sair Jogo']
truques_lista=[]
idade= 0
dict_nivel = {'nivel_fome':10, 'nivel_sujeira':5, 'nivel_brincadeira':5, 'nivel_sono':1}

#imagens
dinosauro = Image.open('dinosauro.jpg')
gato = Image.open('gato.jpg')
cachorro = Image.open('cachorro.jpeg')
A = Image.open('dino/A.png')
B = Image.open('dino/B.png')
C = Image.open('dino/C.png')
D = Image.open('dino/D.png')
E = Image.open('dino/E.png')
F = Image.open('dino/F.png')
O = Image.open('dino/O.png')
I = Image.open('dino/I.png')
U = Image.open('dino/U.png')
R = Image.open('dino/R.png')
S = Image.open('dino/S.png')

class ClassAnimalEstimacao:
    def __init__(self,nome,tipo,idade:int,alimentacao:list,comportamento:dict={'comportamento':'docil','humor':'feliz'}):
        self.nome=nome
        self.tipo =tipo
        self.idade= idade 
        self.alimentacao=alimentacao
        self.comportamento = comportamento

    def dormir(self,tempo,dict_nivel):
        print('ok, vou dormir',tempo,'segundos')
        self.comportamento['humor'] = 'descansado'
        if tempo > 30:
            dict_nivel['nivel_sono'] = dict_nivel['nivel_sono'] -5
            dict_nivel['nivel_brincadeira'] = dict_nivel['nivel_brincadeira'] +5
        elif tempo > 20:
            dict_nivel['nivel_sono'] = dict_nivel['nivel_sono'] -3
            dict_nivel['nivel_brincadeira'] = dict_nivel['nivel_brincadeira'] +3
        elif tempo > 10:
            dict_nivel['nivel_sono'] = dict_nivel['nivel_sono'] -2
            dict_nivel['nivel_brincadeira'] = dict_nivel['nivel_brincadeira'] +2
        else:
            dict_nivel['nivel_sono'] = dict_nivel['nivel_sono'] -1
            dict_nivel['nivel_brincadeira'] = dict_nivel['nivel_brincadeira'] +1
        return dict_nivel
        
    def comer(self,comida,dict_nivel):
        print('nham nham nham')
        if comida == 'racao':
            dict_nivel['nivel_fome'] = dict_nivel['nivel_fome'] -3
            dict_nivel['nivel_sujeira'] = dict_nivel['nivel_sujeira'] +2
        elif comida == 'fruta':
            dict_nivel['nivel_fome'] = dict_nivel['nivel_fome'] -2
            dict_nivel['nivel_sujeira'] = dict_nivel['nivel_sujeira'] +2
        elif comida == 'legume':
            dict_nivel['nivel_fome'] = dict_nivel['nivel_fome'] -2
            dict_nivel['nivel_sujeira'] = dict_nivel['nivel_sujeira'] +2
        elif comida == 'carne':
            dict_nivel['nivel_fome'] = dict_nivel['nivel_fome'] -4
            dict_nivel['nivel_sujeira'] = dict_nivel['nivel_sujeira'] +2
        return dict_nivel
            
        
    def aprender_truques(self,nome_truque):
        print(f'Truque {nome_truque} aprendido')
        try:
            if nome_truque not in self.truque_lista:
                self.truque_lista += [nome_truque]
        except:
            self.truque_lista = [nome_truque]
            
    def brincadeiras(self,truques_lista,dict_nivel):
        print('Adorei bricar de',truques_lista)
   
        dict_nivel['nivel_brincadeira'] = dict_nivel['nivel_sujeira'] -5
        dict_nivel['nivel_sujeira'] = dict_nivel['nivel_sujeira'] +5

        return dict_nivel
            
    def banho(self,dict_nivel):
        print('Que banho gostoso')
   
        dict_nivel['nivel_sujeira'] = dict_nivel['nivel_sujeira'] -4
        dict_nivel['nivel_sono'] = dict_nivel['nivel_sono'] +4

        return dict_nivel

#print("Welcome", flush = True)

print('''Bem vindo, ao Tamagotchi! 

     OBJETIVO: Criar um animal de estimação virtual.
     
     A motivação do brinquedo consiste em cuidar do animalzinho virtual como se fosse real,
     dando-lhe carinho virtual, comida virtual, banho virtual, cuidados virtuais etc..''', flush = True)

print('Podemos comecar? S/N')
start = input()
start=(start.upper())
if start == 'S':
    print('Ok,vamos começar!')
elif start == 'N':
    print('Ok, Jogamos depois!')
else :
    print( 'Favor digitar s para se SIM! Voce deseja comecar a jogar ou N para se não, você não quer jogar agora!')
if start == 'S':
    print('Qual animal de estimação voce gostaria de criar? Cachorro, Gato ou Dinosauro?', flush = True)
    animal_tipo = input()
    animal_tipo=(animal_tipo.upper())
    if animal_tipo == 'GATO':
        display(gato)
        animal_escolhido=1
        time.sleep(0.2)
        print('''Obrigada por me escolher como seu animal de estimação!
        Eu preciso de um nome, como voce quer me chamar?''', flush = True)
        nome= input()
        animal_dict[nome] = animal_tipo
        animal_dict[nome] = ClassAnimalEstimacao(nome,animal_tipo,idade, alimentacao)
    elif animal_tipo == 'CACHORRO':
        display(cachorro)
        animal_escolhido=1
        time.sleep(0.2)
        print('''Obrigada por me escolher como seu animal de estimação!
        Eu preciso de um nome, como voce quer me chamar?''', flush = True)
        nome= input()
        animal_dict[nome] = animal_tipo
        animal_dict[nome] = ClassAnimalEstimacao(nome,animal_tipo,idade, alimentacao)
    elif animal_tipo == 'DINOSAURO':
        display(dinosauro)
        animal_escolhido=1
        time.sleep(0.2)
        print('''Obrigada por me escolher como seu animal de estimação!
        Eu preciso de um nome, como voce quer me chamar?''', flush = True)
        nome= input()
        animal_dict[nome] = animal_tipo
        animal_dict[nome] = ClassAnimalEstimacao(nome,animal_tipo,idade, alimentacao)
    else: 
        raise ValueError('Só é possivel criar: Cachorro, gato ou dinosauro')
    
    
while animal_escolhido==1 and animal_vivo==1:

    if dict_nivel['nivel_fome'] > 5:
        print('humm.. Estou com fome, o que tem para comer?')
        print('Escolha entre', animal_dict[nome].alimentacao)
        comida=input()
        comida=(comida.upper())
        if comida== 'RACAO'or comida== 'fruta'or comida==  'legume'or comida== 'carne':
            dict_nivel= animal_dict[nome].comer(comida,dict_nivel)
        else :
            print('SO gosto de racao, fruta, legume e carne')
        
    if dict_nivel['nivel_sujeira'] > 5:
        print('Estou sujo, pode me dar um banho? S/N')
        res_banho=input()
        res_banho=(res_banho.upper())
        if res_banho == 'S':
            dict_nivel= animal_dict[nome].banho(dict_nivel)
        else:
            print('ok, depois eu tomo banho!')
    
    if dict_nivel['nivel_brincadeira'] > 5:
        if len(truques_lista) > 0:
            print('Vamos brincar de que?')
            truques_lista=input() 
        else:
            print('Eu sei brincar de',truques_lista, 'quer brincar?' )
            quer_brincar=input() 
            quer_brincar=(quer_brincar.upper())
            if quer_brincar == 'S':
                dict_nivel= animal_dict[nome].brincadeiras(truques_lista,dict_nivel)
            else:
                print('ok, depois a gente brinca!')
            
    
    if dict_nivel['nivel_sono'] > 5:
        print('Estou cansado, podemos dormir um pouco?')
        print('Digite quanto tempo quer dormir:')
        tempo=input()
        tempo=int(tempo)
        dict_nivel= animal_dict[nome].dormir(tempo,dict_nivel)
    
    else:
        print('O que voce deseja fazer agora?',lista_desejo)
        
        print('Meu nivel de fome é:',dict_nivel['nivel_fome'])
        print('Meu nivel de sono é:',dict_nivel['nivel_sono'])
        print('Meu nivel de brincadeira é:',dict_nivel['nivel_brincadeira'])
        print('Meu nivel de banho é:',dict_nivel['nivel_sujeira'])
        desejo=input()
        desejo= (desejo.upper())
        if desejo == 'BANHO':
            dict_nivel= animal_dict[nome].banho(dict_nivel)
        elif desejo == 'COMER':
            print('Escolha entre', animal_dict[nome].alimentacao)
            comida=input()
            comida=(comida.upper())
            if comida== 'RACAO'or comida== 'fruta'or comida==  'legume'or comida== 'carne':
                dict_nivel= animal_dict[nome].comer(comida,dict_nivel)
            else :
                print('SO gosto de racao, fruta, legume e carne')
            
        elif desejo == 'DORMIR':
            print('Digite quanto tempo quer dormir:')
            tempo=input()
            tempo=int(tempo)
            dict_nivel= animal_dict[nome].dormir(tempo,dict_nivel)
        elif desejo == 'BRINCAR':
            print('Vamos brincar de que?')
            truques_lista = input()
            dict_nivel= animal_dict[nome].brincadeiras(truques_lista,dict_nivel)
        elif desejo == 'SAIR JOGO':
            animal_vivo=0
        else:
            Print('Não entendi o que voce quer!')
```
